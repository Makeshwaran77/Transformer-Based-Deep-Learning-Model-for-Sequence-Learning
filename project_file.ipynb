```python
import nltk
import re
import string
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from PyPDF2 import PdfReader

nltk.download('punkt')
nltk.download('stopwords')

from nltk.corpus import stopwords
```

```python
def extract_text_from_pdf(file_name):
    reader = PdfReader(file_name)
    text = ""
    for page in reader.pages:
        text += page.extract_text()
    return text

jd_text = extract_text_from_pdf("/content/Job Title_ AI _ ML Engineer.pdf")
resume1 = extract_text_from_pdf("/content/Makeshwaran Resume.pdf")
resume2 = extract_text_from_pdf("/content/Resume - 1.pdf")
resume3 = extract_text_from_pdf("/content/Sibu Resume.pdf")
```

```python
nltk.download('punkt_tab')

stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = nltk.word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]

    return " ".join(tokens)

jd_clean = clean_text(jd_text)
r1_clean = clean_text(resume1)
r2_clean = clean_text(resume2)
r3_clean = clean_text(resume3)
```

```python
documents = [jd_clean, r1_clean, r2_clean, r3_clean]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)
```

```python
similarity_scores = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:]).flatten()

print("Resume 1 Match Score:", round(similarity_scores[0]*100, 2), "%")
print("Resume 2 Match Score:", round(similarity_scores[1]*100, 2), "%")
print("Resume 3 Match Score:", round(similarity_scores[2]*100, 2), "%")
```

```python
best_resume = np.argmax(similarity_scores) + 1
print("\nBest Matching Resume: Resume", best_resume)
```
